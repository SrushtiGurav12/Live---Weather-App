# Onelab — AI Fitness Assessment

## Problem
Reconcile platform transactions with bank settlements and identify mismatches.

## Goal
Detect gaps such as:
- Timing differences
- Rounding errors
- Duplicates
- Missing / orphan records

## Assumptions

1. Each transaction has a unique ID (except duplicates we inject)
2. Settlements occur 1–2 days after transaction
3. Matching is primarily based on transaction ID
4. Refunds are represented as negative amounts
5. Single currency (no FX complications)


```python
import pandas as pd

def generate_data():
    transactions = pd.DataFrame({
        "id": ["t1", "t2", "t3", "t4", "t5", "t5"],  # duplicate
        "amount": [100.00, 200.00, 300.00, 150.25, 99.99, 99.99],
        "timestamp": pd.to_datetime([
            "2024-01-28",
            "2024-01-29",
            "2024-01-30",
            "2024-01-31",
            "2024-01-31",
            "2024-01-31"
        ])
    })

    settlements = pd.DataFrame({
        "id": ["t1", "t2", "t3", "t4", "t6"],
        "amount": [100.00, 200.00, 300.00, 150.24, -50.00],
        "settlement_date": pd.to_datetime([
            "2024-01-29",
            "2024-01-30",
            "2024-02-01",  # next month case
            "2024-02-01",
            "2024-01-31"
        ])
    })

    return transactions, settlements

transactions, settlements = generate_data()

transactions, settlements
```

```python
def reconcile(transactions, settlements):

    # Detect duplicates
    duplicates = transactions[transactions.duplicated(subset=["id"], keep=False)]

    # Merge
    merged = pd.merge(
        transactions,
        settlements,
        on="id",
        how="outer",
        suffixes=("_txn", "_set")
    )

    # Unmatched
    unmatched_txns = merged[merged["amount_set"].isna()]
    unmatched_settlements = merged[merged["amount_txn"].isna()]

    # Rounding differences
    merged["diff"] = merged["amount_txn"] - merged["amount_set"]
    rounding_issues = merged[merged["diff"].abs() > 0.01]

    # Totals
    total_txn = transactions["amount"].sum()
    total_set = settlements["amount"].sum()

    return {
        "duplicates": duplicates,
        "unmatched_txns": unmatched_txns,
        "unmatched_settlements": unmatched_settlements,
        "rounding_issues": rounding_issues,
        "total_txn": total_txn,
        "total_set": total_set,
        "difference": total_txn - total_set
    }

result = reconcile(transactions, settlements)
```

```python
print("=== SUMMARY ===")
print("Total Transactions:", result["total_txn"])
print("Total Settlements:", result["total_set"])
print("Difference:", result["difference"])

print("\n=== DUPLICATES ===")
display(result["duplicates"])

print("\n=== UNMATCHED TRANSACTIONS ===")
display(result["unmatched_txns"])

print("\n=== UNMATCHED SETTLEMENTS ===")
display(result["unmatched_settlements"])

print("\n=== ROUNDING ISSUES ===")
display(result["rounding_issues"])
```

```python
# Duplicate test
assert not result["duplicates"].empty

# Next month settlement test
assert any(settlements["settlement_date"].dt.month == 2)

# Rounding issue test
assert abs(result["difference"]) > 0

# Orphan refund test
assert any(result["unmatched_settlements"]["amount_set"] < 0)

print("All tests passed ✅")
```
